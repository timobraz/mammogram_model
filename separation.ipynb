```python
import pandas as pd
```

```python
df = pd.read_csv("data/description.csv")
df
```

```python
#get a list of path names where class is not NORM (i.e. cancer) and list for non canerous
paths_norm=df[df['Class'] == 'NORM']['Path'].tolist()
paths=df[df['Class'] != 'NORM']['Path'].tolist()

paths[:5]
paths_norm[:5]
```

```python
#write these files into a separate directory called cancerous
import shutil
import os
os.mkdir('data/cancerous')
os.mkdir('data/normal')
for path in paths:
    shutil.copy("./data/images/"+path, 'data/cancerous')
for path in paths_norm:
    shutil.copy("./data/images/"+path, 'data/normal')
```
